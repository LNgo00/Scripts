```python
import pandas as pd
from datetime import datetime, timedelta

historical_data_df = pd.read_csv('historicalData2020_2021.csv')
bbpro_df = pd.read_csv('datatops_bbpro_2021_07_24.csv')
bbgem_df = pd.read_csv('datatops_bbgems_2021_07_24.csv')
```

```python
bbpro_slugs = bbpro_df['slug'].values.tolist()
bbgem_slugs = bbgem_df['slug'].values.tolist()

bb_slugs = bbpro_slugs + bbgem_slugs
```

```python

list_proprojects_slugs = ['kusama','ultra','helium','theta-token','binancecoin','thorchain','okb','metadium','cardano','klay-token',
                'gemini-dollar','melon','reserve-rights-token','origintrail','unibright','true-usd','harmony','quant-network','uquid-coin']

list_gemprojects_slugs = ['force-protocol','lympo','degenerator','wownero','herocoin','lcx','fedoracoin','presearch','duckdaodime','radium',
                         'ferrum-network','troy','zap','ghost-by-mcafee','robonomics-network','yflink','roobee','enigma','elysia','cashbet-coin']

list_projects_slugs = list_proprojects_slugs + list_gemprojects_slugs


def create_datetime_list(datetime_from, num_weeks):
    list_datetime_result = []
    for i in range(0,num_weeks):
        datetime_result = datetime_from + timedelta(i*7)
        datetime_result_str = datetime_result.strftime("%Y-%m-%d")
        list_datetime_result.append(datetime_result_str)
    return list_datetime_result

datetime_from = datetime(year=2020,month=1,day=4)
res = create_datetime_list(datetime_from,8)
print(res)
#print(historical_data_df.head)

def create_custom_df_using_historical(list_projects,list_datetime_result,historical_data_df):
    
    dataframe_slugs = historical_data_df[historical_data_df['slug'].isin(list_projects)]
    dataframe_res = dataframe_slugs[dataframe_slugs['date'].isin(list_datetime_result)]
    
    return dataframe_res

result = create_custom_df_using_historical(bb_slugs,res,historical_data_df)
print(result)
result.to_csv('custom_historical.csv')
```
